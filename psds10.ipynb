# LOOPING

**FOR LOOPING**

```python
# program untuk menghtiung jumlah semua angka dalam sebuah list
L1 = [2,4,5,7,9]
jumlah = 0

#iterasi dalam list
for i in L1:
  jumlah = jumlah + i
print(jumlah)
```

The range() function

```python
# Program untuk melakukan iterasi di dalam list melalui indeks
minuman = ['teh', 'boba', 'kopi', 'susu']
p = len(minuman)
#iterasi di dalam list melalui indeks
for i in range(p):
  print(minuman[i])
```

for loop with else

```python
#contoh for loop menggunakan else
for i in range(p):
  print(minuman[i])
else:
  print('elemen sudah habis')
```

**WHILE LOOP**

```python
#program untuk menjumlahkan semua
#bilangan asli hingga suku ke-n
#jum =1+2+3+...+n

n=10
jumlah=0
i=1
while i<=n:
  jumlah=jumlah+i
  i=i+1
print(jumlah)
```

while loop with else

```python
jum=0
i=1
while i<=n:
  jum=jum+i
  i=i+1
else:
  print(jum)
```

```python
def mahal_diskon(harga):
    if harga >= 500000:
      diskon = harga*0.4
      harga_diskon = harga-diskon
      return 'mahal', 'kamu dapat diskon 40%, harga yang perlu dibayarkan adalah:' + str(harga_diskon)
    elif harga >= 250000:
      diskon = harga*0.2
      harga_diskon = harga-diskon
      return 'mahal', 'kamu dapat diskon 20%, harga yang perlu dibayarkan adalah:' + str(harga_diskon)
    elif harga >= 100000:
      diskon = harga*0.1
      harga_diskon = harga-diskon
      return 'mahal', 'kamu dapat diskon 10%, harga yang perlu dibayarkan adalah:' + str(harga_diskon)
    else:
      return 'murah', 'kamu tidak mendapatkan diskon, harga yang perlu dibayarkan adalah:' + str(harga)
```
